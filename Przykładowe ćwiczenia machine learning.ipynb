# Breast Cancer

```python
from sklearn.datasets import load_breast_cancer

cancer = load_breast_cancer()
X_train, X_test, y_train, y_test = train_test_split(cancer.data, cancer.target, stratify=cancer.target, random_state=66)

training_accuracy = []
test_accuracy = []
neighbors_settings = range(1, 11)

for n_neighbors in neighbors_settings:
    clf = KNeighborsClassifier(n_neighbors=n_neighbors)
    clf.fit(X_train, y_train)
    #Dokładność zestawu uczącego:
    training_accuracy.append(clf.score(X_train, y_train))
    #Dokładność uogólniania:
    test_accuracy.append(clf.score(X_test, y_test))
    
plt.plot(neighbors_settings, training_accuracy, label="dokładność danych uczących")
plt.plot(neighbors_settings, test_accuracy, label="dokładność danych testowych")
plt.ylabel("Dokładność")
plt.xlabel("n_neighbors")
plt.legend()
```

```python
from sklearn.neighbors import KNeighborsRegressor

X, y =mglearn.datasets.make_wave(n_samples=40)
#Split danych na dane uczące i testowe
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=0)
#Utworzenie modelu i wzięcie pod uwagę 3 sąsiadów
reg = KNeighborsRegressor(n_neighbors=3)
#Dopasowanie modelu metodą "fit" danymi uczącymi i celami uczącymi
reg.fit(X_train, y_train)
```

```python
print("Prognozy dot. zestawu testowego:\n{}".format(reg.predict(X_test)))
```

```python
print(reg.score(X_test, y_test))
```

```python
fig, axes = plt.subplots(1, 3, figsize=(15, 4))
#Wykres od -3 do 3 i 1000 punktów danych:
line = np.linspace(-3, 3, 1000).reshape(-1,1)

for n_neighbors, ax in zip([1, 3, 9], axes):
    #podział prognozy na 1, 3 i 9 sąsiadów:
    reg = KNeighborsRegressor(n_neighbors=n_neighbors)
    #Dopasowanie modelu:
    reg.fit(X_train, y_train)
    ax.plot(line, reg.predict(line))
    ax.plot(X_train, y_train, '^', c=mglearn.cm2(0), markersize=8)
    ax.plot(X_test, y_test, 'v', c=mglearn.cm2(1), markersize=8)
    
    #Opis wykresu:
    ax.set_title("{} sąsiad(ów)\n wynik uczenia: {:.2f} wynik testu: {:.2f}".format(n_neighbors, reg.score(X_train, y_train),
                                                                                   reg.score(X_test, y_test)))
    ax.set_xlabel("Cecha")
    ax.set_ylabel("Cel")
    
axes[0].legend(["Prognozy modelu", "Dane uczące/wynik", "Dane testowe/wynik"], loc="best")
```

# Boston Housing model liniowy:

```python
X, y = mglearn.datasets.load_extended_boston()
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=0)
lr = LinearRegression().fit(X_train, y_train)
```

```python
print("Zestaw uczący: {:.2f}".format(lr.score(X_train, y_train)))
print("Zestaw testowy: {:.2f}".format(lr.score(X_test, y_test)))
```

# ----------------------------------------------------------------------------------

## Model regresji grzbietowej

```python
from sklearn.linear_model import Ridge
ridge = Ridge().fit(X_train, y_train)
print("Zestaw uczący: {:.2f}".format(ridge.score(X_train, y_train)))
print("Zestaw testowy: {:.2f}".format(ridge.score(X_test, y_test)))
```

### Dokładność modelu może zostać dopasowana parametrem alpha, który domyślnie wynosi 1.0

### Jeśli zostanie zwiększony do 10:

```python
ridge10 = Ridge(alpha=10).fit(X_train, y_train)
print("Zestaw uczący: {:.2f}".format(ridge10.score(X_train, y_train)))
print("Zestaw testowy: {:.2f}".format(ridge10.score(X_test, y_test)))
```

### Zmniejszenie alpha do 0.1 powoduje małe ograniczenie współczynników, model jest teraz bardzo podobny do zwykłej regresji liniowej:

```python
ridge01 = Ridge(alpha=0.1).fit(X_train, y_train)
print("Zestaw uczący: {:.2f}".format(ridge01.score(X_train, y_train)))
print("Zestaw testowy: {:.2f}".format(ridge01.score(X_test, y_test)))
```

```python
plt.plot(ridge.coef_, 's', label="Ridge dla alpha=1")
plt.plot(ridge10.coef_, '^', label="Ridge dla alpha=10")
plt.plot(ridge01.coef_, 'v', label="Ridge dla alpha=0.1")
plt.plot(lr.coef_, 'o', label="Linear Regression")
plt.xlabel("Wskaźnik współczynnika")
plt.ylabel("Wielkość współczynnika")
plt.hlines(0, 0, len(lr.coef_))
plt.ylim(-10, 10)
plt.legend()
```

## Model Lasso

```python
from sklearn.linear_model import Lasso
lasso = Lasso().fit(X_train, y_train)
print("U: {:.2f}".format(lasso.score(X_train, y_train)))
print("T: {:.2f}".format(lasso.score(X_test, y_test)))
print("Liczba użytych cech: {}".format(np.sum(lasso.coef_ != 0)))
```

```python
plt.plot(lasso.coef_, 's', label="Lasso alpha=1")
plt.xlabel("Wskaźnik współczynnika")
plt.ylabel("Wielkość współczynnika")
plt.ylim(-10, 10)
plt.legend()
```

```python
#Zwiększenie parametru max_iter i zmniejszenie alpha:
lasso001 = Lasso(alpha=0.01, max_iter=100000).fit(X_train, y_train)
print("U: {:.2f}".format(lasso001.score(X_train, y_train)))
print("T: {:.2f}".format(lasso001.score(X_test, y_test)))
print("Liczba użytych cech: {}".format(np.sum(lasso001.coef_ != 0)))
```

```python
lasso00001 = Lasso(alpha=0.0001, max_iter=100000).fit(X_train, y_train)
print("U: {:.3f}".format(lasso00001.score(X_train, y_train)))
print("T: {:.3f}".format(lasso00001.score(X_test, y_test)))
print("Liczba użytych cech: {}".format(np.sum(lasso00001.coef_ != 0)))
```

```python
# wykres o wielu wariantach uogólnienia (alpha)

plt.plot(lasso.coef_, 's', label="Lasso alpha=1")
plt.plot(lasso001.coef_, '^', label="Lasso alpha=0.01")
plt.plot(lasso00001.coef_, 'v', label="Lasso alpha=0.00001")
plt.plot(ridge01.coef_, 'o', label="Ridge aplha=0.1")
plt.legend(ncol=1, loc=(0, 1.05))
plt.ylim(-25,25)
plt.xlabel("Wskażnik współczynnika")
plt.ylabel("Wielkość współczynnika")
```

```python
from sklearn.tree import DecisionTreeClassifier
cancer = load_breast_cancer()
X_train, X_test, y_train, y_test = train_test_split(cancer.data, cancer.target, stratify=cancer.target, random_state=42)
tree = DecisionTreeClassifier(random_state=0)
tree.fit(X_train, y_train)
print("Dokładność w zestawie uczącym: {:.3f}".format(tree.score(X_train, y_train)))
print("Dokładność w zestawie testowym: {:.3f}".format(tree.score(X_test, y_test)))
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.datasets import make_moons

X, y = make_moons(n_samples=100, noise=0.25, random_state=3)
X_train, X_test, y_train, y_test = train_test_split(X, y ,stratify=y, random_state=42)
forest = RandomForestClassifier(n_estimators=5, random_state=2)
forest.fit(X_train, y_train)
```

```python
fig, axes = plt.subplots(2, 3, figsize=(20, 10))
for i, (ax, tree) in enumerate(zip(axes.ravel(), forest.estimators_)):
    ax.set_title("Drzewo {}".format(i))
    mglearn.plots.plot_tree_partition(X_train, y_train, tree, ax=ax)
    mglearn.plots.plot_2d_separator(forest, X_train, fill=True, ax=axes[-1, -1], alpha=.2)
    axes[-1, -1].set_title("Las Fosowy")
    mglearn.discrete_scatter(X_train[:, 0], X_train[:, 1], y_train)
```

```python
X, y = make_blobs(centers=4, random_state=8)
y = y % 2
mglearn.discrete_scatter(X[:, 0], X[:, 1], y)
plt.xlabel("Cecha 0")
plt.ylabel("Cecha 1")
```

```python
#Klasyfikacja w modelu liniowym:
from sklearn.svm import LinearSVC
linear_svm = LinearSVC().fit(X, y)
mglearn.plots.plot_2d_separator(linear_svm, X)
mglearn.discrete_scatter(X[:, 0], X[:, 1], y)
plt.xlabel("Cecha 0")
plt.ylabel("Cecha 1")
```

```python
#Wizualizacja trójwymiarowa
X_new = np.hstack([X, X[:, 1:] ** 2])
from mpl_toolkits.mplot3d import Axes3D, axes3d
figure = plt.figure()
ax = Axes3D(figure, elev=-152, azim=-26, auto_add_to_figure=False)
mask = y == 0
ax.scatter(X_new[mask, 0], X_new[mask, 1], X_new[mask, 2], c='b', cmap=mglearn.cm2, s=60, edgecolor='k')
ax.set_xlabel("cecha0")
ax.set_ylabel("cecha1")
ax.set_zlabel("cecha1 ** 2")
figure.add_axes(ax)
```

```python
linear_svm_3d = LinearSVC().fit(X_new, y)
coef, intercept = linear_svm_3d.coef_.ravel(), linear_svm_3d.intercept_
#Ustawianie granicy decyzyjnej:
figure = plt.figure()
ax = Axes3D(figure, elev=-152, azim=-26, auto_add_to_figure=False)
xx = np.linspace(X_new[:, 0].min() - 2, X_new[:, 0].max() + 2, 50)
yy = np.linspace(X_new[:, 1].min() - 2, X_new[:, 1].max() + 2, 50)
XX, YY = np.meshgrid(xx, yy)
ZZ = (coef[0] * XX + coef[1] * YY + intercept) / -coef[2]
figure.add_axes(ax)
ax.plot_surface(XX, YY, ZZ, rstride=8, cstride=8, alpha=0.3)
ax.scatter(X_new[mask, 0], X_new[mask, 1], X_new[mask, 2], c='b', cmap=mglearn.cm2, s=60)
ax.scatter(X_new[~mask, 0], X_new[~mask, 1], X_new[~mask, 2], c='r', marker='^', cmap=mglearn.cm2, s=60)
ax.set_xlabel=("cecha0")
ax.set_ylabel=("cecha1")
ax.set_zlabel("cecha0 ** 2")
```

```python
ZZ = YY ** 2
dec = linear_svm_3d.decision_function(np.c_[XX.ravel(), YY.ravel(), ZZ.ravel()])
plt.contourf(XX, YY, dec.reshape(XX.shape), levels=[dec.min(), 0, dec.max()], cmap=mglearn.cm2, alpha=0.6)
mglearn.discrete_scatter(X[:, 0], X[:, 1], y)
plt.xlabel("Cecha 0")
plt.ylabel("Cecha 1")
```
